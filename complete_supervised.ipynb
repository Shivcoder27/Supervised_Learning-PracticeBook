# K-nearest Neighbor (work on non linear data, and for both regreesion and classification )

```python
# It work as by counting value of k which is no. of nearest neighbors and separate them which
# is avaible in large quantity the predicted point also become like thata
```

```python
# it uses two method to measure distance
# 1. Manhattan distance formula
# 2. Euclidean Distance formula
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("Social_Network_Ads1.csv")
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x="Age",y = "EstimatedSalary",data = data ,hue = "Purchased")
plt.show()
```

```python
x = data.iloc[:,:-1]
y = data["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
knn.score(x_train,y_train)
```

# TO avoid overfitting 

```python
# for i in range(1,30):
#     knn1 = KNeighborsClassifier(n_neighbors=i)
#     knn1.score(x_train,y_train)
#     print(i,knn1.score(x_train,y_train)*100,knn1.score(x_test,y_test)*100)


for i in range(1, 30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)  # Train the model
    train_accuracy = knn.score(x_train, y_train) * 100
    test_accuracy = knn.score(x_test, y_test) * 100
    print(f"k={i}: Train Accuracy = {train_accuracy:.2f}%, Test Accuracy = {test_accuracy:.2f}%")
```

```python
# here best value is k = 3
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = knn)
plt.show()
```

# Hyper-parameter Tuning

```python
# # Model Parameter are configuration variables that are internal to the model, and a model learns them on its own 
# ex - in linear model u y = mx +c  ..........here m and c are parameter
```

```python
# # Hyperparameter are those parameters that are expilictly defined by the user to control
# the learning process
#  the best value can be determined either by the rule of thumb or by trial and error


# it is default parameters of model ex DecisionTreeClassifier(x ,y ,z) 
```

```python
# we can't tune these hyperparameter manually so we have 
# 1.GridSearchCV
# 2.RandomizedSearchCV
```

```python
data = pd.read_csv("Position_Salaries.csv")
data
```

```python
x = data.iloc[:,:-1]
y = data["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state= 42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor() #hyper parameter in this func ex - max_dept,splitter,criterion etc
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
df = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [i for i in range(2, 20)],
}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid=df, cv=5)

# Fit the model
gd.fit(x_train, y_train)
```

```python
gd.best_params  #it will suggest what will be best hyper parameter for model 
```

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions = df,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params
```
